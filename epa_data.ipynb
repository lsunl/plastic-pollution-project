```python
import pandas as pd
import numpy
```

```python
# read files and put in separate dataframes
all_files = ["epa_data/1 Materials Generated in the U.S. Municipal Waste Stream, 1960 to 2015 (in tons).csv", 
            "epa_data/2 Materials Recycled and Composted in the U.S. Municipal Waste Stream, 1960 to 2015 (in tons).csv",
           "epa_data/3 Materials Combusted with Energy Recovery in the U.S. Municipal Waste Stream, 1960 to 2015 (in tons).csv",
           "epa_data/4 Materials Landfilled in the U.S. Municipal Waste Stream, 1960 to 2015 (in tons).csv"]

li = []

for filename in all_files:
    df = pd.read_csv(filename)
    df1 = df.iloc[[0,1,5,6,9],:]
    df2 = df1.T
    df2.columns = ['paper_paperboard', 'glass', 'metals', 'plastic', 'wood']
    if filename == all_files[0]:
        df2["waste_type"] = "generated"
    elif filename == all_files[1]:
        df2["waste_type"] = "recycled"
    elif filename == all_files[2]:
        df2["waste_type"] = "combusted"
    else:
        df2["waste_type"] = "landfilled"
    frame = df2.iloc[1:]
    li.append(frame)
```

```python
# individually clean df and drop nans

# generated waste
gw = li[0].iloc[:-5]
# recycled waste
rw = li[1]
# combusted waste
cw = li[2]
# landfilled
lw = li[3]
```

```python
# concatenate the dfs into a master df
masterdf = pd.concat([gw,rw,cw,lw], ignore_index=False)
masterdf.rename_axis('year').reset_index()
```
