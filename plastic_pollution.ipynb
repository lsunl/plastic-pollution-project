```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import our data into pandas from CSV
plastic_string = 'Resources/data.csv'
plastic_df = pd.read_csv(plastic_string)

plastic_df.head()
```

```python
plastic_df.columns
```

```python
plastic_df.dropna
```

```python
# Create a group based on the values in the 'maker' column
plastic_gen = plastic_df.groupby('plastic_waste_generation_kg_day7')

# Count how many times each maker appears in our group
plastic_day = plastic_gen['plastic_waste_generation_kg_day7'].count()
```

```python
# Create a bar chart based off of the group series from before
country_chart = plastic_day.plot(kind='bar')

# Set the xlabel and ylabel using class methods
country_chart.set_xlabel("Country")
country_chart.set_ylabel("Plastic Waster Generation (kg)")


plt.show()
plt.tight_layout()
```
